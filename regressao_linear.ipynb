```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

Importando data1.txt

```python
df_1 = pd.read_csv('./data1.txt', names=['population', 'profit'])
df_1
```

Plot do gráfico

```python
df_1.plot(x='population', y='profit', kind='scatter')
```

Realizando a regressão linear

```python
X = np.array(df_1['population']).reshape(-1, 1)
y = np.array(df_1['profit'])

LR = LinearRegression()

LR.fit(X, y)
y_pred = LR.predict(X)

plt.scatter(X, y, color="black", marker='x')
plt.plot(X, y_pred, color="blue", linewidth=1)

plt.xlabel('population')
plt.ylabel('profit')
plt.show()
```
